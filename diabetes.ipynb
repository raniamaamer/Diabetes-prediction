```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Calculer la matrice de corrélation
correlation_matrix = df.corr()

# Afficher la matrice sous forme de heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Matrice de corrélation")
plt.show()

# Trouver les paires d'attributs les plus corrélées (hors la diagonale)
corr_pairs = correlation_matrix.unstack().sort_values(ascending=False)
corr_pairs = corr_pairs[corr_pairs < 1]  # Exclure les corrélations de 1 (diagonale)
strongest_corr = corr_pairs.head(5)  # Les 5 paires les plus corrélées

print("Les paires d'attributs les plus corrélées :\n", strongest_corr)
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
# Séparer les variables explicatives (X) et la cible (y)
X = df.drop(columns=['Outcome'])  
y = df['Outcome']

# Appliquer SelectKBest pour sélectionner les 3 meilleures caractéristiques
selector = SelectKBest(score_func=f_classif, k=3)  # f_classif = ANOVA F-test pour classification
X_new = selector.fit_transform(X, y)

# Récupérer les scores des attributs
scores = selector.scores_

# Créer un DataFrame pour afficher les résultats
feature_scores = pd.DataFrame({'Feature': X.columns, 'Score': scores})
feature_scores = feature_scores.sort_values(by='Score', ascending=False)

# Afficher les 3 meilleurs attributs
print("🔝 Les 3 meilleurs attributs les plus corrélés avec la variable cible :")
print(feature_scores.head(3))
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
```

```python
# Définir X (variables explicatives) et y (cible)
X = df.drop(columns=['Outcome'])  # Toutes les colonnes sauf la cible
y = df['Outcome']

# Modèle de base pour RFE
model = RandomForestClassifier(random_state=42)

# Initialiser RFE pour sélectionner les 3 meilleures caractéristiques
rfe = RFE(estimator=model, n_features_to_select=3)
rfe.fit(X, y)

# Récupérer les meilleures caractéristiques
selected_features = X.columns[rfe.support_]

# Afficher les résultats
print("🔝 Les 3 meilleurs attributs sélectionnés par RFE :")
print(selected_features.tolist())
```
